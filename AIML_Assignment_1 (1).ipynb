```python
#Part 1 : Visualize the datasets or data using Matplotlib Builtin function Graphs
```

```python
import matplotlib.pyplot as plt
people = ['ANN','BRANDON','CHEM','DAVID','EMILY','ASHOK','RAM','RISK',]
age = [18,35,65,34,54,65,51,67]
weight=[50,55,60,65,70,75,80,85]
height=[190,185,180,175,175,180,185,190]
plt.scatter(weight,height)
plt.title("the relationship between the height and weight")
plt.ylabel("height in centimeters")
plt.xlabel("weight in kgs")
plt.show()
```

```python
#Generate bar plot
plt.figure (figsize=(12,4))
plt.title("people's weight in kgs",fontsize=16,fontstyle='italic')
plt.bar(x=people,height=weight, width=0.6,color='orange',edgecolor='k',alpha=0.6)
plt.xlabel("people",fontsize=15)
plt.xticks(fontsize=14,rotation=30)
plt.yticks(fontsize=14)
plt.ylabel("weight (in kgs)",fontsize=15)
plt.show
```

```python
#Generate Histogram
import numpy as np
plt.figure(figsize=(7,5))
plt.hist(weight,color='red',edgecolor='k', alpha=0.8,bins=5)
plt.title("histogram of patient weight",fontsize=20)
plt.xlabel("weight in kgs", fontsize=16)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.show
```
